# 우량주
- 우량주는 업종에 따라, 기업의 성숙도에 따라 기준이 다름.
- 가치주와는 다른 개념
- 가치주는 주식 가격이 중요, 우량주는 기업 자체의 수익성과 안전성



## ROA(자산대비이익), ROE(자본대비이익)
- ROA: Return on Asset. 
- ROA = 당기순이익(Net Income) / 평균 총자산(Total Assets, Averaged) = 당기순이익 /((기초총자산 + 기말총자산)/2)
- ROA가 0.1이라면  "자산: 이익 = 100 : 10"
