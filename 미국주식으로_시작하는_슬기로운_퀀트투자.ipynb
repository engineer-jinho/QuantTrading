## 데이터 불러오기

```python
df = fs.get_price('AAPL', start_date='2020-01-01', end_date='2020-12-31')

#단순 차트
fs.draw_chart(df, right='AAPL')

#테이블
df
```

### 일반차트 vs RSI 차트

```python
#RSI 차트
fs.rsi(df, w=14)
fs.draw_chart(df, left='rsi', right='AAPL')
```

### RSI 기준으로 사고 팔 시기를 표현

```python
fs.indicator_to_signal(df, factor='rsi', buy=40, sell=60)
fs.position(df)
fs.draw_chart(df, left='rsi', right='position_chart')
```

### Evaluate

```python
fs.evaluate(df, cost=.001)
fs.draw_chart(df, left='acc_rtn_dp', right='AAPL')
```

## 최종평가

```python
fs.performance(df, rf_rate=0.01)
```

## 결론

```python
df_apple = fs.get_price('AAPL', start_date='2020-01-01', end_date='2020-12-31')

fs.rsi(df_apple, w=14)
fs.indicator_to_signal(df_apple, factor='rsi', buy=60, sell=40)

fs.position(df_apple)
fs.evaluate(df_apple, cost=.001)
fs.performance(df_apple, rf_rate=0.01)

fs.draw_chart(df_apple, left='rsi', right='position_chart')
fs.draw_chart(df_apple, left='acc_rtn_dp', right='AAPL')
```
